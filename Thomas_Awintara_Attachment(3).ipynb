```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy as cp
import cartopy.crs as ccrs
import cartopy.feature as cf
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from pathlib import Path
import warnings
import pandas as pd
import numpy as np
warnings.filterwarnings('ignore')
```

```python
dat = pd.read_fwf('GMet_location_avgSI.txt', names = ['Station', 'Longitude', 'Latitude', 'St'])
#dat
```

```python

fig=plt.figure(figsize=(10,9))                         ###### Specifying the size of the figure
ax = plt.axes(projection = ccrs.PlateCarree())       ###### Specifying the type of geopatial plot
ax.add_feature(cf.COASTLINE,alpha=0.8)
ax.add_feature(cf.BORDERS, color='violet')
ax.add_feature(cf.RIVERS)
ax.add_feature(cf.LAND)
ax.add_feature(cf.OCEAN)
ax.add_feature(cf.LAKES, color='blue')
ax.set_extent([-3.5,1.2,11.4,4.5])                   #### setting the map boundaries

ax.add_feature(cf.STATES, alpha= 0.5)               ####  adding territorial boundaries
ax.plot(dat.Longitude,
       dat.Latitude,
        'ro',                                       ##### plotting the longitudes and latitudes of the station
       ms=7,
        color = 'orange')

s_stations = np.asarray(dat.Station)
for longitude, latitude, name in zip(dat.Longitude, dat.Latitude, s_stations):
    if name in ['Yendi']:
        ax.text(longitude - .05, latitude - .15,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Zuarungu']:
        ax.text(longitude + .10, latitude - .15,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Navrongo']:
        ax.text(longitude -.10, latitude + .00,
                name,
                va='center',
                ha='right', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Akuse']:
        ax.text(longitude -.25, latitude - .15,
                name,
                va='center',
                ha='left', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Akim_Oda']:
        ax.text(longitude +.10, latitude + .15,
                name,
                va='center',
                ha='right', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Koforidua']:
        ax.text(longitude +.20, latitude + .15,
                name,
                va='center',
                ha='right', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Tema']:
        ax.text(longitude +.20, latitude - .20,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Accra']:
        ax.text(longitude -.15, latitude + .10,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Takoradi']:
        ax.text(longitude +.20, latitude - .20,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    elif name in ['Damongo']:
        ax.text(longitude +.20, latitude - .20,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
    else:
        ax.text(longitude + .09, latitude + .12,
                name,
                va='center',
                ha='center', transform=ccrs.Geodetic(), fontweight='bold',fontsize = '12')
        ax.set_xticks([-3.2,-2.2,-1.2,0,1.2], crs=ccrs.PlateCarree())
ax.set_yticks([11,10,9,8,7,6,5], crs=ccrs.PlateCarree())
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)

```
